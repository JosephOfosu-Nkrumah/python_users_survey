```python
import pandas as pd
import matplotlib.pyplot as plt # library for visualization
from matplotlib.ticker import PercentFormatter #converts values into percentage format
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default
```

```python
Python_Users = pd.read_excel("C:/Users/Joseph Ofosu-Nkrumah/Desktop/New folder/python_users.xlsx") #read in the Python_Users survey data file
```

```python
Python_Users #the data frame, which we'll use to create the combo plot. The two charts will share the x-axis which is Year.
# The bar chart will have the number of Participants on the primary x-axis(on the lefthandside of the chart)
# The line chart will have the Python Users in percentages on a secondary y-axis(on the righthandside of the chart)
```

```python
# Creating the combination chart. Here we use a figure with axes.
# The first axes or subplot is the bar chart, the second is the line chart. 
fig, ax = plt.subplots(figsize = (10, 7))

#creating and styling the bar chart
ax.bar(Python_Users["Year"],
       Python_Users["Participants"],
       color = ['turquoise','yellow','purple','green','orange','violet','brown','blue'])
ax.set_ylabel("Number of Participants", 
              weight = "bold")
ax.set_xlabel("Year", weight = "bold")
ax.tick_params(axis = "y", 
               width = 2, 
               labelsize = "large")
ax1 = ax.twinx()
#changing the secondary y-axis to display percentages on a scale from 0% to 100%
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
#creating and styling the line chart
ax1.plot(Python_Users["Year"],
         Python_Users["Python Users"], 
         color = "#b60000", 
         marker = "D")
ax1.set_ylabel("Python Users in %", 
               color = "#b60000", 
               weight = "bold")
ax1.tick_params(axis = "y", 
                colors = "#b60000", 
                width = 2, 
                labelsize = "large")
ax.set_title("Survey of Python Users (2012 - 2019)", fontsize = "14", weight = "bold")
plt.savefig("Survey of Python Users (2012 - 2019).png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```
